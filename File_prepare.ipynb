```python
# process dataframe
### get columns with 'stage'
subdf = data[data['description'].str.contains('stage')].reset_index(drop=True)
#if문 안쓰고 str.contains('stage')형태로 간단하게 저장
stages = list(subdf['description'])
#list형태로 간단하게 저장
onsets = list(subdf['onset'])
```

```python
### Compile the stages
new_stages = [stages[0]]
new_onsets = [onsets[0]]
for idx,x in enumerate(stages[0:len(stages)-1]):
    cur = x #current
    nxt = stages[idx+1] #next
    if cur != nxt:
        new_stages.append(nxt)
        new_onsets.append(onsets[idx+1])
compDF = pd.DataFrame(list(zip(new_stages, new_onsets)),columns=['stage','onset'])
```

```python
# Split the arrays
split_arrays = []
split_dfs = []
startind = int(compDF.onset[0]*256)
dfstartind = 0
onsets = list(compDF.onset)
stages = list(compDF.stage)
ind = 0
while ind < len(compDF)-1:
    cur = stages[ind]
    nxt = stages[ind+1]
    if nxt == 'Sleep stage ?': 
        # Sleep stage에 들어갔을때 첫인덱스 0으로 시작해서 ? 시작 전 인덱스 까지 값들을 각각 df 형태 np형태로 저장 시키는 부분
        # Subset and store the array
        endind = int(onsets[ind+1]*256)
        subarray = eeg_array[:,np.arange(startind, endind)]
        split_arrays.append(subarray)
        # Subset the dataframes and find the times/ind
        nsubdf = compDF[dfstartind:ind+1].reset_index(drop=True)
        first_time = nsubdf.onset[0]
        nsubdf['real_time'] = nsubdf['onset'] - first_time
        real_indices = [int(x*256) for x in nsubdf['real_time']]
        nsubdf['real_index'] = real_indices
        split_dfs.append(nsubdf)
        # 첫 인덱스를 ?시작 시점의 +1을 한 시점으로 dfstartind / startind를 저장 시켜서 while문으로 다시 돌리는 형태 
        if ind+1 < len(compDF)-1:
            startind = int(onsets[ind+2]*256)
            dfstartind = ind+2
    ind+=1
```
